**Case Study on Numpy & Pandas**

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("auto-mpg.csv")
df.head()
```

##**NumPy Problems**

1. Basic Array Operations Convert the mpg column into a NumPy array and calculate:


* The mean, median, and standard
deviation of mpg.
* The number of cars with mpg greater than 25.

```python
# Convert mpg column to NumPy array
mpg_array = df['mpg'].to_numpy()

# Calculate statistics
mpg_mean = np.mean(mpg_array)
mpg_med = np.median(mpg_array)
mpg_std = np.std(mpg_array)

print("Mean MPG:", mpg_mean)
print("Median MPG:", mpg_med)
print("Standard Deviation of MPG:", mpg_std)

# Count cars with mpg > 25
cars = np.sum(mpg_array > 25)

print("Number of cars with MPG greater than 25:", cars)
```

2. Filtering

Using NumPy, filter all cars with more than 6 cylinders.

Return the corresponding car_name as a list.

```python
# Convert required columns to NumPy arrays
cylinders_array = df['cylinders'].to_numpy()
car_names_array = df['car name'].to_numpy()

# Create a boolean mask for cars with more than 6 cylinders
mask = cylinders_array > 6

# Apply mask to car names
cars_more_than_6_cyl = car_names_array[mask]

# Convert result to a Python list
cars_more_than_6 = cars_more_than_6_cyl.tolist()

# Print result
length = len(cars_more_than_6)
print(f"Total cars with more than 6 cylinders: {length}\n")
print(cars_more_than_6)
```

3. Statistical Analysis

Compute the 25th, 50th, and 75th percentiles of the weight column using NumPy.

```python
# Convert weight column to NumPy array
weight_array = df['weight'].to_numpy()

# Compute percentiles
percent_25 = np.percentile(weight_array, 25)
percent_50 = np.percentile(weight_array, 50)
percent_75 = np.percentile(weight_array, 75)

print(f"25th percentile of weight: {percent_25}")
print(f"50th percentile (median) of weight: {percent_50}")
print(f"75th percentile of weight: {percent_75}")
```

```python
percentiles = np.percentile(weight_array, [25, 50, 75])
print("25th, 50th, 75th percentiles of weight:", percentiles)
```

4. Array Manipulation

Convert the acceleration column into a NumPy array and normalize its values
(scale between 0 and 1).


```python

acceleration_array = df['acceleration'].to_numpy()

# Normalize using min-max scaling
acceleration_norm = (acceleration_array - np.min(acceleration_array)) / (np.max(acceleration_array) - np.min(acceleration_array))

# Print first 10 normalized values
print(acceleration_norm[:10])
```

5. Broadcasting

Increase all horsepower values by 10% and store the updated values in a new
NumPy array. Handle missing data (if any) by replacing it with the mean of the
column before applying the increase

```python
import numpy as np

# Convert 'horsepower' to NumPy array and replace '?' with np.nan
hp = np.array([float(x) if x != '?' else np.nan for x in df['horsepower'].to_numpy()])

# Replace NaN with mean and increase by 10%
hp_upd = np.where(np.isnan(hp), np.nanmean(hp), hp) * 1.10

# Print first 10 updated values
print(hp_upd[:10])
```

```python

# Check if any NaN is left in hp_upd
print("Any missing values left?", np.isnan(hp_upd).any())
```

6. Boolean Indexing

Find the average displacement of cars with an origin of 2 (Europe) using NumPy
indexing.


```python
# Convert the columns to NumPy arrays
displacement = df['displacement'].to_numpy()
origin = df['origin'].to_numpy()

# Create a boolean mask for European cars (origin == 2)
mask = origin == 2

# Use the mask to select displacement of European cars
displacement_europe = displacement[mask]

# Compute the average
avg_displacement_europe = np.mean(displacement_europe)

print("Average displacement of European cars:", avg_displacement_europe)
```

7. Matrix Operations

Create a 2D NumPy array containing the columns mpg, horsepower, and weight.
Compute the dot product of this matrix with a given vector [1, 0.5, -0.2].


```python
# Convert columns to numeric arrays
mpg = df['mpg'].to_numpy()
weight = df['weight'].to_numpy()
hp = np.array([float(x) if x != '?' else np.nan for x in df['horsepower'].to_numpy()])
hp = np.where(np.isnan(hp), np.nanmean(hp), hp)

# Create 2D matrix
matrix = np.column_stack((mpg, hp, weight))

# Vector for dot product
vec = np.array([1, 0.5, -0.2])

# Compute dot product
dot_result = np.dot(matrix, vec)

# Show first 10 results
print(dot_result[:10])
```

9. Correlation

Compute the Pearson correlation coefficient between mpg and weight using

10. Conditional Aggregates

Calculate the mean mpg for cars grouped by the number of cylinders using NumPy
techniques.

```python
import numpy as np

mpg = df['mpg'].to_numpy()
cyl = df['cylinders'].to_numpy()

for c in np.unique(cyl):
    mean_mpg = np.mean(mpg[cyl == c]) #find average mpg for this cylinder type
    print(f"Cylinders: {c}, Mean MPG: {mean_mpg:.2f}")
```

Find all cars manufactured in 1975 with a weight less than 3000. Return the

3. Handling Missing Data

Identify if there are any missing values in the dataset. Replace missing values in the
horsepower column with the column's median.

```python
# Check for missing values in each column
missing_values = df.isnull().sum()
print("Missing values in each column:\n", missing_values)
```

```python
# Note: The horsepower column may have missing values marked as '?',
# which is not counted as NaN by Pandas.

# Check for '?' specifically in horsepower
missing_hp_count = (df['horsepower'] == '?').sum()
print("Number of missing values in horsepower (marked as '?'):", missing_hp_count)
```

```python
df['horsepower'] = pd.to_numeric(df['horsepower'], errors='coerce')

# Replace NaN values with median (safe way)
median_hp = df['horsepower'].median()
df['horsepower'] = df['horsepower'].fillna(median_hp)

# Confirm
print("Missing values in horsepower after replacement:", df['horsepower'].isnull().sum())


# Show rows filled with median
print(f"\nRows where horsepower was replaced with median ({median_hp}):")
display(df[df['horsepower'] == median_hp][['car name', 'horsepower']])
```

10. **Finding Anomalies**

Identify potential outliers in the mpg column using the Interquartile Range (IQR)
method. Specifically:
* Calculate the IQR for mpg.
2
* Define outliers as values less than Q1 - 1.5 * IQR or greater than Q3 +
1.5 * IQR.
* Create a DataFrame of cars classified as outliers, displaying car_name, mpg,
and model_year.

```python
# Q1, Q3, and IQR
Q1 = df['mpg'].quantile(0.25)
Q3 = df['mpg'].quantile(0.75)
IQR = Q3 - Q1

# lower and upper bounds for outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Filter rows where mpg is an outlier
outliers = df[(df['mpg'] < lower_bound) | (df['mpg'] > upper_bound)][['car name', 'mpg', 'model year']]

display(outliers)
```
