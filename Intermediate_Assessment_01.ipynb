2.3 Find the item with the highest total quantity sold across all
orders.

```python
# Dictionary to store total quantity per item
qty = {}

# Add quantities for each item
for s in sales_dat:
    qty[s["item"]] = qty.get(s["item"], 0) + s["quantity"]

# Find the item with highest quantity
top_item = max(qty, key=qty.get)

# Print the result
print("Item with highest total quantity sold:", top_item, "\nQuantity:", qty[top_item])
```



* 4.1 Write a custom iterator class OrderIterator to iterate through the
orders, yielding one order at a time.

* 4.2 Create a dictionary grouping all orders by customer using a
dictionary comprehension.



```python
# 4.1
class OrderIterator:
    def __init__(self, orders):
        self.o = orders  # store orders
        self.i = 0       # start index

    def __iter__(self):
        return self

    def __next__(self):
        if self.i < len(self.o):
            order = self.o[self.i]
            self.i += 1
            return order
        else:
            raise StopIteration
```

```python
# Example usage
it = OrderIterator(sales_dat)
for order in it:
    print(order)
```

```python

it = OrderIterator(sales_dat)

# Collect orders in a list
orders_list = [order for order in it]

# Convert to DataFrame for table format
df_orders = pd.DataFrame(orders_list)

# Display the table
df_orders
```

```python
# 4.2
#Group orders by customer using dictionary comprehension
orders_by_customer = {cust: [o for o in sales_dat if o["customer"] == cust]
                      for cust in set(o["customer"] for o in sales_dat)}

# Print the grouped dictionary
for cust, orders in orders_by_customer.items():
    print("\nCustomer:", cust)
    for order in orders:
        print(order)
```
